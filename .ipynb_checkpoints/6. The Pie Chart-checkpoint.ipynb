# The Pie Chart

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
# The minimalistic pie chart
plt.pie(slices)
plt.show()
```

```python
# Adding labels and title
plt.pie(slices, labels=labels)
plt.title('Popularity of Top Programming Languages')
plt.show()
```

```python
# Changing colors
colors = ['blue', 'green', 'black', 'orange', 'yellow']  # Sequence in correspondance with slices and labels
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, colors=colors)
plt.show()
```

```python
# Adding outline to each slice
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
# Accentuating a slice
explode = [0, 0, 0, 0.1, 0]
# It detaches the 4th slice (Python in our case) from the pie at a distance equivalent to 10 percent of pie radius
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode)
plt.show()
```

```python
# Adding shadow (gives a 3D impression)
explode = [0, 0, 0, 0.1, 0]
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode, shadow=True)
plt.show()
```

```python
# Setting starting angle of the first slice
explode = [0, 0, 0, 0.1, 0]
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode, shadow=True, startangle=90)
# It'll position the first slice (JavaScript) at 90 degrees and the other slices follow in sequence
plt.show()
```

```python
# Adding percentage to each slice
explode = [0, 0, 0, 0.1, 0]
plt.title('Popularity of Top Programming Languages')
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')
plt.savefig('plots/pie chart.png', dpi=300)
plt.show()
```
