# Bar Chart From Pandas DataFrame

```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
plt.style.use('fivethirtyeight')
```

### Plotting number of users of most popular languages

```python
df = pd.read_csv('dataset/data.csv', index_col='Responder_id')
df.head(5)
```

```python
languages_counter = Counter()
for element in df['LanguagesWorkedWith']:
    languages_counter.update(element.split(';'))
```

```python
languages_counter
# Automatically sorted with highest count key first
```

```python
# Picking only top 15 languages
common_languages = languages_counter.most_common(15)
common_languages
```

```python
languages = []
popularity = []
for element in common_languages:
    languages.append(element[0])
    popularity.append(element[1])
languages.reverse()
popularity.reverse()
```

```python
# Plotting a horizontal bar chart
# For horizontal bar chart use: barh(yaxis, xaxis)
plt.barh(languages, popularity)
plt.ylabel('Programming Languages')
plt.xlabel('No. of Users')
plt.title('Most Popular Languages')
plt.tight_layout()
plt.show()
```
