# Filling Line Plots

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataset/salaries.csv')
df.head(5)
```

```python
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
```

```python
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Filling between zero and a python devs line
plt.fill_between(ages, py_salaries)
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Reducing the opacity of fill
plt.fill_between(ages, py_salaries, alpha=0.3)
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Filling between median salary of all devs and python devs
median_salary = dev_salaries.median()
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary)
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Adding conditions to fill (filling only when python dev salary > median salary)
median_salary = dev_salaries.median()
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary, where=(py_salaries > median_salary), interpolate=True)
# Interpolate=True fills some area which doesn't fulfil the condition to give a smoother look
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Filling area between median and python dev salaries with different color above and below the median
median_salary = dev_salaries.median()
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary, where=(py_salaries > median_salary), interpolate=True)
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary, where=(py_salaries <= median_salary), interpolate=True)
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Adding labels to the fills
median_salary = dev_salaries.median()
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary, where=(py_salaries > median_salary),interpolate=True,
                label='Above Median')
plt.fill_between(ages, py_salaries, alpha=0.3, y2=median_salary, where=(py_salaries <= median_salary), interpolate=True,
                label='Below Median')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
# Filling area between all dev salaries and python dev salaries with different colors
median_salary = dev_salaries.median()
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, label='Python Devs')
plt.fill_between(ages, py_salaries, alpha=0.3, y2=dev_salaries, where=(py_salaries > dev_salaries), interpolate=True,
                label="Above Avg")
plt.fill_between(ages, py_salaries, alpha=0.3, y2=dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True,
                label="Below Avg")
plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```
