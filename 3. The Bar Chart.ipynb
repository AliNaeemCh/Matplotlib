### Plotting Multiple Bars
For plotting multiple bars, we have to position them manually:
1. **The default single bar width is 0.8. Divide it by number of bars in the chart, i-e for 2 bars set width = 0.4.**
2. The xticks are positioned at the center of the bar or at position half of bar width, which means for a single bar where width = 0.8, the tick is at 0.4. Similarly, for 2 bars, the tick will be at 0.2. **So, our goal is to set bar_width/2 as the center of all our bars i-e to both its left and right width of 0.4 each should be present, giving a total of 0.8. In this way, we can properly position the xticks at the center of our bars.**
3. **To avoid bars being superimposed on eachother, we'll add an offset to the x-axis of each bar such that as one bar ends, the other starts.**

```python
def calc_offsets (total_bars, bar_width):
    total_width = 0.8
    offsets = []
    center = bar_width/2
    o = center - total_width/2
    for i in range(total_bars):
        offsets.append(o)
        o += bar_width
    return offsets
```

```python
# Testing function for 2 bars. Center will be 0.2. One bar should be at offset -0.2 and other at 0.2
total_bars = 2
calc_offsets(total_bars, 0.8/total_bars)
```

```python
def plot_bars (bars_data, x_axis=None, legends=None, title=None, xlabel=None, ylabel=None, colors=None, save=None):
    total_bars = len(bars_data)
    bars_width = 0.8/total_bars
    x_indexes = np.arange(len(bars_data[0]))
    if x_axis == None:
        x_axis = x_indexes
    if colors == None:
        colors = [None for _ in range(total_bars)]
    offsets = calc_offsets(total_bars, bars_width)
    for i in range(total_bars):
        plt.bar(x_indexes + offsets[i], bars_data[i], color =  colors[i], width=bars_width)
    plt.xticks(ticks=x_indexes, labels=x_axis)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    if legends != None:
        plt.legend(legends)
    if save:
        plt.savefig(save['path'], dpi=save.get('dpi', 100))
    plt.show()
    return
```

```python
plot_bars(
            bars_data=[dev_y, py_dev_y, js_dev_y],
            x_axis=ages,
            legends=['All Devs', 'Python Devs', 'JS Devs'],
            title='Median Salary (USD) by Age',
            xlabel='Age',
            ylabel='Salary (USD)',
            save={'path': 'plots/multiple bars.png', 'dpi':300}
         )
```
