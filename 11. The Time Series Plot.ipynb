# The Time Series Plot

```python
from matplotlib import pyplot as plt
from matplotlib import dates
import pandas as pd
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
df = pd.read_csv('dataset/time series.csv')
df.head(5)
```

```python
print('Data type of Date column is:', df['Date'].dtype)
```

```python
# Converting Date column to Date time format
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
price_close = df['Close']
price_date = df['Date']
price_date
```

```python
# The minimalistic time series plot
plt.plot_date(price_date, price_close)
plt.title('Digital Currency Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tight_layout()
plt.show()
```

```python
# Refining
plt.plot_date(price_date, price_close, linestyle='-')
plt.gcf().autofmt_xdate()              # This makes the dates slanted so they don't look congested. gcf = get current figure
plt.title('Digital Currency Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tight_layout()
plt.show()
```

```python
# Formatting the date
plt.plot_date(price_date, price_close, linestyle='-')
plt.gcf().autofmt_xdate()
date_format = dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.title('Digital Currency Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tight_layout()
plt.show()
```
