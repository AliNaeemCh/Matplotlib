# The Scatter Plot

```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
df = pd.read_csv('dataset/trending.csv')
df.head(5)
```

```python
views = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.scatter(views, likes)
plt.title('Trending Videos Stats')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# Changing style, opacity, and size
plt.scatter(views, likes, s=50, c='green', edgecolor='black', linewidth=1, alpha=0.8)
plt.title('Trending Videos Stats')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# Using colormap to change color of each data point by like/dislike ratio
plt.scatter(views, likes, s=50, c=ratio, cmap='viridis', edgecolor='black', linewidth=1, alpha=0.8)
plt.colorbar().set_label('Like/Dislike Ratio')
plt.title('Trending Videos Stats')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.tight_layout()
plt.show()
```

```python
# Applying log scale to both axes for better visibility
plt.scatter(views, likes, s=50, c=ratio, cmap='viridis', edgecolor='black', linewidth=1, alpha=0.8)
plt.colorbar().set_label('Like/Dislike Ratio')
plt.title('Trending Videos Stats')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```
